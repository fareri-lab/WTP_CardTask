```python
x = ELAdata['BRCS']
y = ELAdata['QUIC_parental_predictability']

plot1 = sns.regplot(x='BRCS',y='QUIC_parental_predictability',data = ELAdata,color='red')

r,p = scipy.stats.pearsonr(x, y)
print('Pearsons correlation: %.3f' % r)
print('p value: %.4f' % p)
```
