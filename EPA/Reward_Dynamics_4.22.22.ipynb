```python
#change directory to where the .csv files containing the data are stored on your computer
os.chdir('/Users/dfareri/Dropbox/Dominic/Adelphi_Teaching/talia/')

#read in csv file and set display options
VS_long = pd.read_csv(r'VS_activation_zstat_long.csv', encoding='utf-8-sig')
pd.set_option("display.max_rows", None, "display.max_columns", None)

#print the dataframe
print(VS_long)
```

```python
#Use seaborn to create timeseries plot

#settings for plot
sns.set(style="white",font="Arial")
sns.set_context("paper", font_scale=2.0)
plt.figure(figsize=(8,6))
sns.set(style="ticks")

#create using lineplot
test = sns.lineplot(data=VS_long, x="TR", y="Reward Response", hue="Condition",ci=68,palette = "mako")

#set axis labels and title
test.set_ylabel("Reward Response(z-stat)",fontsize=18)
test.set_title("VS Timecourse",fontsize=18)
test.set_xlabel("TR",fontsize=18)
test.tick_params(labelsize=18)

# modify legend text
plt.setp(test.get_legend().get_texts(), fontsize='15') 
 
# modify legend title
plt.setp(test.get_legend().get_title(), fontsize='15')

#show plot
plt.show()

test.figure.savefig('VS_timecourse.jpeg')
```
