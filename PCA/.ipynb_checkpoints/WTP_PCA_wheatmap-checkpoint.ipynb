```python
#extract column of participant ids and drop from dataframe
ids = wtpdata['subid']
wtpdata = wtpdata.drop(['subid', 'subnum', 'type', 'phase','balanced','dem_AGE', 'dem_SEX', 'dem_GENDER', 'dem_ETHNICITY', 'dem_RACE', 'avg_spent_social', 'avg_spent_nonsocial',  
        'total_spent_social', 'total_spent_nonsocial',  'prop_social',  'prop_nonsocial', 'moresocial',  'prop_moresocial',  'spent'  ,'total_bonus', 'PC1','PC2','PC3'], axis=1)
#store id column in variable called "measures"
measures = wtpdata.columns
```

```python
#add id column as the index in the dataframe
data = pd.DataFrame(data=wtpdata)
data.index = ids
print(data)
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(wtpdata)
x = pd.DataFrame(x, columns=measures)
print(x)
```

```python
from sklearn.decomposition import PCA
pcamodel = PCA(n_components=3)
pca = pcamodel.fit_transform(x)
pca.shape
```

```python
pcamodel.explained_variance_ 
pcamodel.explained_variance_ratio_
```

```python
plt.bar(range(1,len(pcamodel.explained_variance_ )+1),pcamodel.explained_variance_ )
plt.ylabel('Explained variance')
plt.xlabel('Components')
plt.plot(range(1,len(pcamodel.explained_variance_ )+1),
         np.cumsum(pcamodel.explained_variance_),
         c='red',
         label="Cumulative Explained Variance")
plt.legend(loc='upper left')
```

```python
plt.plot(pcamodel.explained_variance_ratio_)
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()

#PCA1 is at 0 in xscale
```

```python
measures = ['Autism Quotient','BISBAS - BIS','BISBAS - Reward Responsiveness', 'BIS - Fun Seeking','BIS - Drive',
            'Resilience', 'Depression', 'COVID - Experience', 'COVID - Impact', 'COVID - Stress', 'COVID - Threat', 
            'IRI - Fantasy', 'IRI - Empathy', 'IRI - Personal Distress', 'IRI - Perspective Taking', 'Social Anxiety',
           'Social Anxiety - Fear', 'Social Anxiety - Avoidance', 'Mood', 'Social Support', 'Solitude', 'Self Esteem',
           'Anxiety', 'Loneliness','Unified Motives - Affiliation', 'Unified Motives - Intimacy']
```

```python
sns.set(font_scale=1.5)
fig, ax = plt.subplots(figsize=(25,35)) 
ax = sns.heatmap(pcamodel.components_,
                 cmap='RdBu_r',
                 yticklabels=[ "PCA"+str(x) for x in range(1,pcamodel.n_components_+1)],
                 xticklabels=list(measures),
                 cbar_kws={"orientation": "horizontal"})
ax.set_aspect("equal")

```
